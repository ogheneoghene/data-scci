```python
plt.hist(df['salary_in_usd'], bins=10)
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Salary')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['employment_type'].value_counts().plot(kind='bar')
plt.xlabel('Employment Type')
plt.ylabel('Count')
plt.title('Distribution of Employment Types')
plt.show()
```

```python
plt.scatter(df['work_year'], df['salary_in_usd'])
plt.xlabel('Work Year')
plt.ylabel('Salary (USD)')
plt.title('Salary vs. Work Year')
plt.show()
```

```python
avg_salary_by_exp_level = df.groupby('experience_level')['salary_in_usd'].mean()
print(avg_salary_by_exp_level)
```

```python
median_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].median()
print(median_salary_by_company_size)
```

```python
plt.figure(figsize=(10, 6))
df['job_title'].value_counts().plot(kind='bar')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.title('Distribution of Job Titles')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['remote_ratio'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Remote Work Ratio')
plt.show()
```

```python
avg_salary_by_job_title = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_salary_by_job_title.plot(kind='bar')
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Job Title')
plt.show()
```

```python
top_countries = df['company_location'].value_counts().nlargest(10).index

plt.figure(figsize=(8, 6))
df[df['company_location'].isin(top_countries)]['company_location'].value_counts().plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Company Locations')
plt.show()
```
